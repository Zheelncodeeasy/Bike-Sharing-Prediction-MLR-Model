## Business Goal:
#### You are required to model the demand for shared bikes with the available independent variables. It will be used by the management to understand how exactly the demands vary with different features. They can accordingly manipulate the business strategy to meet the demand levels and meet the customer's expectations. Further, the model will be a good way for management to understand the demand dynamics of a new market. 

### Data Dictionary : 
#### 
- instant: record index
- dteday : date
- season : season (1:spring, 2:summer, 3:fall, 4:winter)
- yr : year (0: 2018, 1:2019)
- mnth : month ( 1 to 12)
- holiday : weather day is a holiday or not (extracted from http://dchr.dc.gov/page/holiday-schedule)
- weekday : day of the week
- workingday : if day is neither weekend nor holiday is 1, otherwise is 0.
+ weathersit : 
    - 1: Clear, Few clouds, Partly cloudy, Partly cloudy
    - 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
    - 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
    - 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog
- temp : temperature in Celsius
- atemp: feeling temperature in Celsius
- hum: humidity
- windspeed: wind speed
- casual: count of casual users
- registered: count of registered users
- cnt: count of total rental bikes including both casual and registered

```python
def add_counts(ax):
       for p in ax.patches: 
           ax.annotate(f'{int(p.get_height())}',
                       (p.get_x() + p.get_width() / 2., p.get_height()), 
                       ha='center', va='bottom', 
                       xytext=(0, 5), 
                       textcoords='offset points', 
                      fontsize=10, color='black')

def plot_catcols(columns):
    for col in columns:
        plt.figure(figsize=(10,7))
        ax1 = sns.countplot(x=col,data=bikedata)
        add_counts(ax1)
        plt.title(col.capitalize())
        plt.show()
        
plot_catcols(cat_cols)
        
```

## Split Train and Test Data

```python
# Splitting the Data into Training and Testing Sets
df_train, df_test = train_test_split(bikedata,train_size=0.7,random_state=100)

print(df_train.shape)
print(df_test.shape)
```

### Scaling the data for Uniformity

```python
# scaling data using minmaxscaler (Range between 0 and 1)

# Instantiate an object
scaler = MinMaxScaler()

# list of numeric variables
numeric_cols = ['temp','hum','windspeed','cnt']

# Fit on data
df_train[numeric_cols] = scaler.fit_transform(df_train[numeric_cols])
df_train.describe()
```

```python
#heatmap

plt.figure(figsize=(25,15))
sns.heatmap(df_train.corr(),annot = True, cmap = 'YlGnBu')
plt.show()
```

## Training the model

```python
# X_train, y_train

y_train = df_train.pop('cnt')
X_train = df_train
X_train.shape
```

## Building the model using RFE (Recursive Feature Elimination)

```python
# Building the model using an automated approach using RFE that will help select the top 15 features

lm = LinearRegression()

lm.fit(X_train,y_train)

rfe = RFE(lm,n_features_to_select = 15)
rfe = rfe.fit(X_train,y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
# These are the columns that are selected by rfe
rfesupport_cols = X_train.columns[rfe.support_]
rfesupport_cols
```

```python
# These are the columns that are not selected by rfe
X_train.columns[~rfe.support_]
```

- Now, we have selected 15 independent variables using RFE. 
- Moving on, we can use manual approach by removing independent variables based on their p values and VIF.

## Building model using statsmodel with RFE selected variables

```python
# Creating train dataframe for variables selected by RFE

X_train_rfe = X_train[rfesupport_cols]
```

```python
X_train_sm = sm.add_constant(X_train_rfe)

# create model
lr = sm.OLS(y_train,X_train_sm)

# fit
lr_model = lr.fit()

# params
lr_model.summary()
```

```python
vif = pd.DataFrame()
vif['Features'] = X_train_sm.columns
vif['VIF'] = [variance_inflation_factor(X_train_sm.values,i) for i in range(X_train_sm.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
vif = vif.sort_values(by = 'VIF', ascending = False)
vif
```

```python
# We will drop Fall which is insignificant as it has high p value with high vif 

X_train_sm2 = X_train_sm.drop(['Fall'], axis = 1)
```

### Rebuilding the model again 

```python
# add a constant variable

X_train_sm2 = sm.add_constant(X_train_sm2)
```

```python
# create model
lr = sm.OLS(y_train,X_train_sm2)

# fit
lr_model = lr.fit()

# params
lr_model.summary()
```

```python
vif = pd.DataFrame()
vif['Features'] = X_train_sm2.columns
vif['VIF'] = [variance_inflation_factor(X_train_sm2.values,i) for i in range(X_train_sm2.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
vif = vif.sort_values(by = 'VIF', ascending = False)
vif
```

```python
X_train_sm2.shape
```

## Model Building Observation

1. Using RFE approach, we selected the top 15 independent variables. Then we proceeded with manual approach and build the model with 15 variables.
2. Using statsmodel, we build the model where we got the adjusted R square as 84.3 % 
3. We noticed that variable Fall had high pf and vif value, so we dropped it and build the model again.
4. In the final model, the adjusted R square remained the same indicating variable Fall was a insignificant variable.
5. After building the final model, all the pvalues for independent variables were below 0.05 and vif was below 5.
6. The F-statistic for the final model had a very small p-value (8.70e-192), indicating a highly significant fit among the variables.
.

## Residual Analysis

```python
y_train_pred = lr_model.predict(X_train_sm2)
```

```python
residual = y_train - y_train_pred
residual
```

```python
y_train_pred
```

```python
cols = X_train_sm2.columns
cols
```

### Assumption 1 - Error terms are normally distributed

```python
fig = plt.figure(figsize=(15,6))
sns.set_style("darkgrid")
sns.distplot(residual, bins = 20)
fig.suptitle('Error Terms', fontsize = 20)                  # Plot heading 
plt.xlabel('Errors', fontsize = 18)    
plt.show()
```

### Error Terms are centered towards mean showing normal distrubution. Hence respecting the assumption.

### Assumption 2 - Error terms have constant variance (homoscedasticity)  and Linearity

```python
plt.figure(figsize=(15,6))
sns.set_style("darkgrid")
plt.scatter(lr_model.fittedvalues, lr_model.resid)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.title('Residuals vs Fitted Values')
plt.show()
```

### Random scatter of data points around the zero line explains the assumption of homoscedasticity being respected. 
### Since the residuals are randomly scattered, there is no visible trend in the plot. This suggests that the relationship between the predictor variables and the target variable is linear, as expected in linear regression.

### Assumption 3 : Multicollinearity

```python
# Multicollinearity
vif_data = pd.DataFrame()
vif_data["feature"] = X_train_sm2.columns
vif_data["VIF"] = [variance_inflation_factor(X_train_sm2.values, i) for i in range(len(X_train_sm2.columns))]
print(vif_data)
```

### VIF values for all the independent variables are below 5. Hence indicating no significant multicollinearity issues.

### Assumption 4 : Error terms are independent of each other 

```python
from statsmodels.stats.stattools import durbin_watson
dw_stat = durbin_watson(lr_model.resid)
print(f'Durbin-Watson statistic: {dw_stat}')
```

### Since the value is 2, we can say there is little to no autocorrelation. Hence, indicating that error terms are independent of each other
